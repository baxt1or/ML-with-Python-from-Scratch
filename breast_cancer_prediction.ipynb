```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/breast_cancer.csv")

df.head(2)
```

```python
diagnosis_count = df["diagnosis"].value_counts()

plt.figure(figsize=(8,6))
diagnosis_count.plot(kind="bar", color=['skyblue', 'salmon'])
```

```python
df.columns[2:]
```

```python
df["diagnosis"] = df["diagnosis"].apply(lambda x: 1 if x=="M" else 0)

df.head(4)
```

```python
plt.figure(figsize=(20,15))
plotnumber=1
for column in df:
    if plotnumber<=30:
        ax = plt.subplot(5,6, plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column)
    plotnumber+=1

plt.tight_layout()
plt.show()
```

```python
df.drop(columns="id", inplace=True)

plt.figure(figsize=(20,12))  
sns.heatmap(data=df.corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5)
```

```python
df.corr()
```

```python
corr_max = df.corr().abs()
mask = np.triu(np.ones_like(corr_max, dtype=bool))
tri_df = corr_max.mask(mask)

to_drop = [x for x in tri_df.columns if any(tri_df[x] > 0.92)]

df.drop(to_drop, axis=1)
```

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X_train, X_test, y_train , y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lres = LogisticRegression()

lres.fit(X_train, y_train)
```

```python
y_train_pred = lres.predict(X_train)
y_test_pred = lres.predict(X_test)
```

```python
train_accuracry_log = accuracy_score(y_train, y_train_pred)
test_accuracry_log = accuracy_score(y_test, y_test_pred)


print(f'Accuracy in Train Set: {train_accuracry_log}')
print(f'Accuracy in Test Set: {test_accuracry_log}')

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)
```

```python
train_accuracry_knn = accuracy_score(y_train, y_train_pred)
test_accuracry_knn = accuracy_score(y_test, y_test_pred)


print(f'Accuracy in Train Set: {train_accuracry_knn}')
print(f'Accuracy in Test Set: {test_accuracry_knn}')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
svc= SVC(probability=True)

parameters = {
    'gamma': [0.0001, 0.001, 0.01, 0.1],
    'C':[0.01, 0.05, 0.5, 0.1, 1,10, 15,20]
}
grid_search = GridSearchCV(svc, parameters)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
svc = SVC(C=15, gamma=0.01, probability=True)
svc.fit(X_train, y_train)
```

```python

train_accuracry_svm = accuracy_score(y_train, svc.predict(X_train))
test_accuracry_svm = accuracy_score(y_test, svc.predict(X_test))


print(train_accuracry_svm)
print(test_accuracry_svm)
```

```python
models = pd.DataFrame({
    "Model":["Logistic Regression", "KNN", "SVM"],
    "Score": [100*round(test_accuracry_log,4), 100*round(test_accuracry_knn,4), 100*round(test_accuracry_svm,4)] 
})

models.sort_values(by = 'Score', ascending = False)
```

```python
import pickle
models = lres
pickle.dump(models, open("breast_cancer.pkl", "wb"))
```

```python
from sklearn import metrics
plt.figure(figsize=(8,5))
models = [
{
    'label': 'LR',
    'model': lres,
},
{
    'label': 'SVM',
    'model': svc,
},
{
    'label': 'KNN',
    'model': knn,
},

]

for m in models:
    model = m['model'] 
    model.fit(X_train, y_train) 
    y_pred=model.predict(X_test) 
    fpr1, tpr1, thresholds = metrics.roc_curve(y_test, model.predict_proba(X_test)[:,1])
    auc = metrics.roc_auc_score(y_test,model.predict(X_test))
    plt.plot(fpr1, tpr1, label='%s - ROC (area = %0.2f)' % (m['label'], auc))

plt.plot([0, 1], [0, 1],'r--')
plt.xlim([-0.01, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1 - Specificity (False Positive Rate)', fontsize=12)
plt.ylabel('Sensitivity (True Positive Rate)', fontsize=12)
plt.title('ROC - Breast Cancer Prediction', fontsize=12)
plt.legend(loc="lower right", fontsize=12)
plt.savefig("roc_breast_cancer.jpeg", format='jpeg', dpi=400, bbox_inches='tight')
plt.show()
```
